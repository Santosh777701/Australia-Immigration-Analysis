```python
import pandas as pd
```

```python
file_path="C:/SQL/Australia_data.csv"
```

```python
df = pd.read_csv(file_path)
```

```python
#checking first 10 rows 
df.head(10)
```

```python
df.columns = df.columns.str.strip()  # Clean column names
print("Data Loaded Successfully!")
df.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
total_immigration_2008 = df['2008'].sum()
print("Total Immigration in 2008:", total_immigration_2008)
```

```python
top_countries = df[['OdName', '2008']].sort_values(by='2008', ascending=False).head(5)
print("Top 5 Countries with Highest Immigration in 2008:")
print(top_countries)
```

```python
# Plotting the graph on the  Immigration Trends Over Years
plt.figure(figsize=(12,6))
df.set_index('OdName')[['1980', '1990', '2000', '2008']].head(10).plot(kind='bar', figsize=(12,6))
plt.title("Immigration Trends Over Years")
plt.ylabel("Number of Immigrants    ")
plt.xlabel("Country")
plt.legend(title="Years")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='OdName', y='2008', data=top_countries, palette="viridis")
plt.title("Top 5 Countries with Highest Immigration in 2008")
plt.xlabel("Country")
plt.ylabel("Number of Immigrants")
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.show()
```

```python
# 3. Distribution of Immigration in 2008
plt.figure(figsize=(10,5))
sns.histplot(df['2008'], bins=30, kde=True, color='blue')
plt.title("Distribution of Immigration Numbers in 2008")
plt.xlabel("Number of Immigrants")
plt.ylabel("Frequency")
plt.show()
```

```python
#corelation heat map
plt.figure(figsize=(12,8))
sns.heatmap(df.iloc[:, 5:].corr(), annot=True, cmap="coolwarm", fmt='.2f', annot_kws={"size": 8}, linewidths=0.5, linecolor='black')
plt.title("Correlation Heatmap of Immigration Data")
plt.show()
```

```python
output_file = "Australia_data_processed.csv"
df.to_csv(output_file, index=False)
print(f"Processed data saved to {output_file}")

print("âœ… Jupyter Notebook Data Analysis Completed Successfully!")
```
